## 4.1 Exercise on Functions
### Task 1 - Unit Conversion Program

```python
# Task 1: Measurement Converter

def convert_length(value, from_unit, to_unit):
    if from_unit == "m" and to_unit == "ft":
        return value * 3.28084
    elif from_unit == "ft" and to_unit == "m":
        return value / 3.28084

def convert_weight(value, from_unit, to_unit):
    if from_unit == "kg" and to_unit == "lbs":
        return value * 2.20462
    elif from_unit == "lbs" and to_unit == "kg":
        return value / 2.20462

def convert_volume(value, from_unit, to_unit):
    if from_unit == "l" and to_unit == "gal":
        return value * 0.264172
    elif from_unit == "gal" and to_unit == "l":
        return value / 0.264172


print("Unit Converter")
print("1. Length")
print("2. Weight")
print("3. Volume")

choice = input("Enter your choice (1/2/3): ")

value = float(input("Enter value: "))
from_unit = input("From unit: ").lower()
to_unit = input("To unit: ").lower()

if choice == "1":
    answer = convert_length(value, from_unit, to_unit)
    print("Converted value =", answer)

elif choice == "2":
    answer = convert_weight(value, from_unit, to_unit)
    print("Converted value =", answer)

elif choice == "3":
    answer = convert_volume(value, from_unit, to_unit)
    print("Converted value =", answer)

else:
    print("Invalid choice")
```

### Task 2 - Mathematical Operations on a List of Numbers

```python
# Task 2: Mathematical Expression

def calculate_sum(numbers):
    return sum(numbers)

def calculate_average(numbers):
    return sum(numbers) / len(numbers)

def find_maximum(numbers):
    return max(numbers)

def find_minimum(numbers):
    return min(numbers)

print("Choose what you want to find:")
print("1. Sum")
print("2. Average")
print("3. Maximum")
print("4. Minimum")


choice = input("Enter your choice (1/2/3/4): ")

numbers = list(map(float, input("Enter numbers separated by spaces: ").split()))

print("\nYou entered:", numbers)

if choice == "1":
    result = calculate_sum(numbers)
    print("Operation selected: Sum")
    print("The sum of the numbers is:", result)

elif choice == "2":
    result = calculate_average(numbers)
    print("Operation selected: Average")
    print("The average of the numbers is:", round(result, 2))

elif choice == "3":
    result = find_maximum(numbers)
    print("Operation selected: Maximum")
    print("The maximum number is:", result)

elif choice == "4":
    result = find_minimum(numbers)
    print("Operation selected: Minimum")
    print("The minimum number is:", result)

else:
    print("Invalid choice. Please run the program again.")
```

```python
# 3. Get Sublist using Slicing

def get_sublist(lst, start, end):
    return lst[start:end+1]

my_list = [1, 2, 3, 4, 5, 6]
sub_list = get_sublist(my_list, 2, 4)

print("Original list 1:", my_list)
print("Start index: 2")
print("End index: 4")
print("The sublist from index 2 to 4 is:", sub_list)

my_list_2 = ['a', 'b', 'c', 'd', 'e']
sub_list_2 = get_sublist(my_list_2, 0, 2)

print("Original list 2:", my_list_2)
print("Start index: 0")
print("End index: 2")
print("The sublist from index 0 to 2 is:", sub_list_2)
```

```python
# 4. Remove First and Last Elements

def remove_first_last(lst):
    return lst[1:-1]

list1 = [1, 2, 3, 4, 5]
result1 = remove_first_last(list1)

print("For the first list:")
print("Original list =", list1)
print("New list after removing first and last elements =", result1)
print()

list2 = ['a', 'b', 'c', 'd', 'e']
result2 = remove_first_last(list2)

print("For the second list:")
print("Original list =", list2)
print("New list after removing first and last elements =", result2)
```

```python
# 5. Get the First n Elements

def get_first_n(lst, n):
    return lst[:n]

list1 = [1, 2, 3, 4, 5, 6, 7, 8]
n1 = 3
result1 = get_first_n(list1, n1)

print("For the first list:")
print("Original list =", list1)
print("n =", n1)
print("New list with first n elements =", result1)
print()

list2 = ['a', 'b', 'c', 'd', 'e']
n2 = 2
result2 = get_first_n(list2, n2)

print("For the second list:")
print("Original list =", list2)
print("n =", n2)
print("New list with first n elements =", result2)
```

```python
# 6. Extract Elements from the End

def get_last_n(lst, n):
    return lst[-n:]

my_list = [1, 2, 3, 4, 5, 6, 7, 8]

print("Original list:", my_list)

n1 = 3
print("When n =", n1)
print("The last", n1, "elements are:", get_last_n(my_list, n1))
print()

n2 = 5
print("When n =", n2)
print("The last", n2, "elements are:", get_last_n(my_list, n2))
print()

n3 = 1
print("When n =", n3)
print("The last", n3, "element is:", get_last_n(my_list, n3))
```

```python
# 7. Extract Elements in Reverse Order

def reverse_skip(lst):
    return lst[-2::-2]

list1 = [1, 2, 3, 4, 5, 6]
print("For the first list:")
print("Original list =", list1)
print("Output =", reverse_skip(list1))
print()

list2 = [10, 20, 30, 40, 50]
print("For the second list:")
print("Original list =", list2)
print("Output =", reverse_skip(list2))
print()

list3 = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
print("For the third list:")
print("Original list =", list3)
print("Output =", reverse_skip(list3))
```

## 4.3 Exercise on Nested List

```python
# Nested List Operations

def flatten(lst):
    result = []
    for item in lst:
        if type(item) == list:
            result += flatten(item)
        else:
            result.append(item)
    return result

def access_nested_element(lst, indices):
    element = lst
    for i in indices:
        element = element[i]
    return element

def sum_nested(lst):
    total = 0
    for item in lst:
        if type(item) == list:
            total += sum_nested(item)
        else:
            total += item
    return total

def remove_element(lst, elem):
    result = []
    for item in lst:
        if type(item) == list:
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result

def find_max(lst):
    flat_list = flatten(lst)
    return max(flat_list)

def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if type(item) == list:
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

def deep_flatten(lst):
    return flatten(lst)

def average_nested(lst):
    flat_list = flatten(lst)
    return sum(flat_list) / len(flat_list)


# Example lists
nested_1 = [[1, 2], [3, 4], [5]]
nested_2 = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
nested_3 = [[1, 2], [3, [4, 5]], 6]
nested_4 = [[1, 2], [3, 2], [4, 5]]
nested_5 = [[1, 2], [2, 3], [2, 4]]
nested_6 = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
nested_7 = [[1, 2], [3, 4], [5, 6]]

print("Nested List Operation")

print("\n1. Flatten a nested list")
print("Original list:", nested_1)
print("Flattened list:", flatten(nested_1))

print("\n2. Access an element from nested list")
print("Original list:", nested_2)
print("Index path: [1, 2]")
print("Element found:", access_nested_element(nested_2, [1, 2]))

print("\n3. Find the sum of all elements in nested list")
print("Original list:", nested_3)
print("Sum of all elements:", sum_nested(nested_3))

print("\n4. Remove an element from nested list")
print("Original list:", nested_4)
print("Element to remove: 2")
print("New list after removing 2:", remove_element(nested_4, 2))

print("\n5. Find the maximum element in nested list")
print("Original list:", nested_3)
print("Maximum element:", find_max(nested_3))

print("\n6. Count occurrences of an element")
print("Original list:", nested_5)
print("Element to count: 2")
print("Number of times 2 appears:", count_occurrences(nested_5, 2))

print("\n7. Deep flatten a nested list")
print("Original list:", nested_6)
print("Fully flattened list:", deep_flatten(nested_6))

print("\n8. Find average of all elements in nested list")
print("Original list:", nested_7)
print("Average of all elements:", average_nested(nested_7))
```

### Problem 4 - Matrix Operations

```python

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

print(f"Given matrices:\nA = \n{A}\nB = \n{B}")

# 1. Prove A.A⁻¹ = I
inverse_A = np.linalg.inv(A)
product_A_invA = np.dot(A, inverse_A)
identity_matrix = np.eye(2)

print(f"\n1. Proving A.A⁻¹ = I:")
print(f"Inverse of A is:\n{inverse_A}")
print(f"A . A⁻¹ =\n{product_A_invA}")
print(f"Identity matrix I =\n{identity_matrix}")
print(f"Is A . A⁻¹ approximately equal to I? {np.allclose(product_A_invA, identity_matrix)}")

# 2. Prove AB ≠ BA
product_AB = np.dot(A, B)
product_BA = np.dot(B, A)

print(f"\n2. Proving AB ≠ BA:")
print(f"A . B =\n{product_AB}")
print(f"B . A =\n{product_BA}")
print(f"Is A . B equal to B . A? {np.array_equal(product_AB, product_BA)}")

# 3. Prove (AB)ᵀ = BᵀAᵀ
product_AB_T = np.transpose(product_AB)
transpose_B = np.transpose(B)
transpose_A = np.transpose(A)
product_BT_AT = np.dot(transpose_B, transpose_A)

print(f"\n3. Proving (AB)ᵀ = BᵀAᵀ:")
print(f"(A . B)ᵀ =\n{product_AB_T}")
print(f"Bᵀ . Aᵀ =\n{product_BT_AT}")
print(f"Is (A . B)ᵀ approximately equal to Bᵀ . Aᵀ? {np.allclose(product_AB_T, product_BT_AT)}")
```

### Solve the System of Linear Equations


```python
# Represent the system AX = C
coeff_matrix = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3, 1, -1]
], dtype=float)

constants = np.array([-1, -3, 9], dtype=float)

# Method 1: Solve directly using np.linalg.solve
solution_solve = np.linalg.solve(coeff_matrix, constants)

# Method 2: Solve using the inverse matrix
coeff_matrix_inv = np.linalg.inv(coeff_matrix)
solution_inv = coeff_matrix_inv @ constants

print("Coefficient matrix A:\n", coeff_matrix)
print("\nConstant vector C:\n", constants)
print("\nSolution using np.linalg.solve:")
print(f"x = {solution_solve[0]}, y = {solution_solve[1]}, z = {solution_solve[2]}")

print("\nSolution using np.linalg.inv(A) @ C:")
print(f"x = {solution_inv[0]}, y = {solution_inv[1]}, z = {solution_inv[2]}")

print("\nDo both methods match?", np.allclose(solution_solve, solution_inv))
```

```python
# Shared data for 1,000,000-element experiments
SIZE = 1_000_000
list_a = [random.random() for _ in range(SIZE)]
list_b = [random.random() for _ in range(SIZE)]

np_a = np.array(list_a)
np_b = np.array(list_b)
```

### 1. Element-wise Addition

```python
# Python lists
start = time.perf_counter()
list_add_result = [a + b for a, b in zip(list_a, list_b)]
end = time.perf_counter()
python_add_time = end - start

# NumPy arrays
start = time.perf_counter()
numpy_add_result = np_a + np_b
end = time.perf_counter()
numpy_add_time = end - start

print(f"Python list addition time: {python_add_time:.6f} seconds")
print(f"NumPy array addition time: {numpy_add_time:.6f} seconds")
print(f"Speedup: {python_add_time / numpy_add_time:.2f}x")
```

### 2. Element-wise Multiplication

```python
# Python lists
start = time.perf_counter()
list_mul_result = [a * b for a, b in zip(list_a, list_b)]
end = time.perf_counter()
python_mul_time = end - start

# NumPy arrays
start = time.perf_counter()
numpy_mul_result = np_a * np_b
end = time.perf_counter()
numpy_mul_time = end - start

print(f"Python list multiplication time: {python_mul_time:.6f} seconds")
print(f"NumPy array multiplication time: {numpy_mul_time:.6f} seconds")
print(f"Speedup: {python_mul_time / numpy_mul_time:.2f}x")
```

### 3. Dot Product

```python
# Python lists
start = time.perf_counter()
list_dot_result = sum(a * b for a, b in zip(list_a, list_b))
end = time.perf_counter()
python_dot_time = end - start

# NumPy arrays
start = time.perf_counter()
numpy_dot_result = np.dot(np_a, np_b)
end = time.perf_counter()
numpy_dot_time = end - start

print(f"Python list dot product time: {python_dot_time:.6f} seconds")
print(f"NumPy array dot product time: {numpy_dot_time:.6f} seconds")
print(f"Speedup: {python_dot_time / numpy_dot_time:.2f}x")
print(f"Are the results close? {np.isclose(list_dot_result, numpy_dot_result)}")
```
